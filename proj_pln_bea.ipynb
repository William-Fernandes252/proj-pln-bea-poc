## Carregar relatórios gerenciais

```
import os
from langchain_community.document_loaders import PyPDFLoader

documents_dir_path = "/content/drive/MyDrive/Colab Notebooks/bea_documents"

# List to hold all loaded documents
all_documents = []

# Check if the directory exists
if os.path.exists(documents_dir_path):
    print(f"Directory found at: {documents_dir_path}")
    # Iterate over files in the directory
    for filename in os.listdir(documents_dir_path):
        if filename.endswith(".pdf"): # Process only PDF files
            pdf_path = os.path.join(documents_dir_path, filename)
            print(f"Loading document: {filename}")
            try:
                loader = PyPDFLoader(pdf_path)
                current_documents = loader.load()
                print(f"Successfully loaded {len(current_documents)} pages from {filename}.")
                all_documents.extend(current_documents)
            except Exception as e:
                print(f"Error loading or reading PDF {filename}: {e}")

    # After loading all documents, assign to 'documents' variable
    documents = all_documents

    if len(documents) > 0:
        print(f"\nTotal documents loaded: {len(documents)} pages.")
        print("\n--- First page content of the first loaded document (partial) ---")
        print(documents[0].page_content[:500] + "...") # Print first 500 characters
        print("------------------------------------")
    else:
        print("No PDF documents were loaded from the directory.")
else:
    print(f"Error: Directory not found at {documents_dir_path}. Please ensure your Google Drive is mounted and the path is correct.")
```

## Configurar RAG (com FAISS)

```
import requests
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain_openai import OpenAIEmbeddings
from langchain_community.vectorstores import FAISS
from langchain_classic.tools.retriever import create_retriever_tool

# 2. Split the document
text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=100)
texts = text_splitter.split_documents(documents)

# 3. Create Embeddings and Vector Store (FAISS instead of Qdrant)
embeddings = OpenAIEmbeddings(model="text-embedding-3-small")
vector_store = FAISS.from_documents(texts, embeddings)

# 4. Create the Retriever
retriever = vector_store.as_retriever()

# 5. Wrap as a Tool for the Agent
retriever_tool = create_retriever_tool(
    retriever,
    name="search_management_reports",
    description="Útil para responder a perguntas envolvendo relatórios gerenciais, governança, estratégia e demonstrações contábeis de fundos de investimento (como o Maxi Renda) e empresas."
)

print("RAG Tool created successfully.")
```

## Configurar agente

```
from langchain_openai import ChatOpenAI
from langchain_tavily import TavilySearch
from langchain_classic.agents import AgentExecutor, create_tool_calling_agent
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_community.chat_message_histories import ChatMessageHistory
from langchain_core.chat_history import BaseChatMessageHistory
from langchain_core.runnables.history import RunnableWithMessageHistory

# 1. Define Tools
search_tool = TavilySearch()
tools = [search_tool, retriever_tool]

# 2. Initialize LLM
llm = ChatOpenAI(model="gpt-5-nano", temperature=0)

# 3. Create Prompt with History Placeholder
# We add MessagesPlaceholder for "chat_history" so the agent can see previous turns
prompt = ChatPromptTemplate.from_messages([
    ("system", "Você é um assistente financeiro especialista em analisar notícias e relatorios gerenciais de fundos de investimento e empresas listadas na B3. Utilize as ferramentas disponíveis para responder às perguntas dos usuários."),
    MessagesPlaceholder(variable_name="chat_history"),
    ("human", "{input}"),
    ("placeholder", "{agent_scratchpad}"),
])

# 4. Create Agent
agent = create_tool_calling_agent(llm, tools, prompt)
agent_executor = AgentExecutor(agent=agent, tools=tools, verbose=True)

# 5. Setup History Management
# This dictionary simulates a database of session histories
store = {}

def get_session_history(session_id: str) -> BaseChatMessageHistory:
    if session_id not in store:
        store[session_id] = ChatMessageHistory()
    return store[session_id]

# 6. Wrap Agent with History
agent_with_chat_history = RunnableWithMessageHistory(
    agent_executor,
    get_session_history,
    input_messages_key="input",
    history_messages_key="chat_history",
)

print("Agent with memory initialized.")
```

## Definir utilitário para chat

```
def chat_loop():
    print("--- Bea - Brazilian Equity Analyzer (digite 'quit' ou 'exit' para sair) ---")
    print("Experimente perguntar: 'Qual é o preço da Taesa?' e depois 'E do Banco do Brasil?'")

    # We define a static session ID for this loop so history is preserved
    session_id = "session_1"

    while True:
        try:
            # Get user input
            user_input = input("\nUsuário: ")

            # Check for exit commands
            if user_input.lower() in ["quit", "exit", "q"]:
                print("Encerrando conversa...")
                break

            if not user_input.strip():
                continue

            # Invoke the agent with the session configuration
            response = agent_with_chat_history.invoke(
                {"input": user_input},
                config={"configurable": {"session_id": session_id}}
            )

            # Print the final answer
            print(f"Bea: {response['output']}")

        except KeyboardInterrupt:
            print("\nEncerrando conversa...")
            break
```
